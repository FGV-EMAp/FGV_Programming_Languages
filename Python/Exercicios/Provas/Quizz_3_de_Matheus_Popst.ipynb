1 - Dadas uma lista qualquer, como a do exemplo abaixo:

```python
lista = [1,2,3,'4',5,6.3,7.4 + 2j,"123",[1,2,3], 93, "98", (1,2,3,'4')]
```

Crie uma função recursiva para imprimir o antecessor e o sucessor de cada elemento de uma lista:
+ Caso este elemento seja inteiro, imprimir o valor;  
+ Caso seja uma string, imprimir o inteiro correspondente;  
+ Caso seja um real, imprima a parte inteira;
+ Caso seja um imaginário, imprima a parte inteira e real;
+ Caso seja uma sequência, aplique as regras acima para cada elemento da sequência.  

```python
def caso(lista):
    for item in lista:
        if type(item)==int:
            print(item)
        elif type(item)==str:
            print(int(item))
        elif type(item)==float:
            print(int(item))
        elif type(item)==complex:
            print(int(item.real))
        elif type(item)==list or type(item)==tuple:
            caso(item)
```

```python
caso(lista)
```

2 - Crie uma função, ou conjunto de funções, para exibir por extenso o valor de qualquer número inteiro até 1000.  
Exemplo:  

    print(imprime_extenso(234))  
    > 'duzentos e trinta e quatro'  

```python
def imprime_extenso(n):
    n = str(n)
    a=len(n)
    N = []
    for i in range(a):
        N.append(int(n[i]))
    if N==[0]:
        return 'zero'
    if N==[1, 0, 0, 0]:
        return 'mil'
    if N==[1, 0, 0]:
        return 'cem'
    if N==[1, 0]:
        return 'dez'
    unidade = {0:'zero', 1:'um', 2:'dois', 3:'tres', 4:'quatro', 5:'cinco', 6:'seis', 7:'sete', 8:'oito', 9:'nove'}
    #classe 1
    if len(N)==1:
        return unidade[N[0]]
    dezena = {2:'vinte', 3:'trinta', 4:'quarenta', 5:'cinquenta', 6:'sessenta', 7:'setenta', 8:'oitenta', 9:'noventa'}
    centena = {1:'cento', 2:'duzentos', 3:'trezentos', 4:'quatrocentos', 5:'quinhentos', 6:'seiscentos', 7:'setecentos', 8:'oitocentos', 9:'novecentos'}
    if int(n)%100==0:
        return centena[N[0]]
    def adolescente(N):
        teen = {11:'onze', 12:'doze', 13:'treze', 14:'quatorze', 15:'quinze', 16:'dezesseis', 17:'dezessete', 18:'dezoito', 19:'dezenove'}
        return teen[N[1]+10]
    #classe 2
    def classe2(N):
        if N[0]==1:
            return adolescente(N)
        elif N[1]==0:
            return dezena[N[0]]
        else:
            return dezena[N[0]] + ' e ' + unidade[N[1]]
    if len(N)==2:
        return classe2(N)
    #classe 3
    if len(N)==3:
        a=N.pop(0)
        if N[0]==0:
            return centena[a] + ' e ' + unidade[N[1]]
        elif N[0]==1 and N[1]==0:
            return centena[a] + ' e dez'
        else:
            return centena[a] + ' e ' + classe2(N)
```

```python
for i in range(0, 1000):
    print(imprime_extenso(i))
```

3- Em alguns lugares é comum lembrar um número do telefone associando seus dígitos a letras. Dessa maneira a expressão MY LOVE significa 69 5683. Claro que existem alguns problemas, uma vez que alguns números de telefone não formam uma palavra ou uma frase e os dígitos 1 e 0 não estão associados a nenhuma letra.  

Sua tarefa é ler uma expressão e encontrar o número de telefone correspondente baseado na tabela abaixo. Uma expressão é composta por letras maiúsculas (A-Z), hifens (-) e os números 1 e 0.  

Letras  ->  Número 
ABC    ->  2 
DEF    ->  3 
GHI    ->  4 
JKL    ->  5 
MNO    ->  6 
PQRS    ->  7 
TUV    ->  8 
WXYZ   ->  9 

Entrada  

A entrada consiste de um conjunto de expressões. Cada expressão está sozinha em uma linha e possui C caracteres, onde 1 ≤ C ≤ 30. A entrada é terminada por fim de arquivo (EOF).  

Saída  

Para cada expressão você deve imprimir o número de telefone correspondente.  

Exemplo de entrada:  

1-HOME-SWEET-HOME  
MY-MISERABLE-JOB  

Saída correspondente:

1-4663-79338-4663  
69-647372253-562  

```python
def telefone(n):
    A=[letra for letra in n]
    Cat = {'1':'1', '0':'0', '-':'-', 'A':'2', 'B':'2', 'C':'2', 'D':'3', 'E':'3', 'F':'3', 'G':'4', 'H':'4', 'I':'4', 'J':'5', 'K':'5', 'L':'5', 'M':'6', 'N':'6', 'O':'6', 'P':'7', 'Q':'7', 'R':'7', 'S':'7', 'T':'8', 'U':'8', 'V':'8', 'W':'9', 'X':'9', 'Y':'9', 'Z':'9'}
    n=''
    for letra in A:
        n+=Cat[letra]
    print(n)
```

```python
telefone('1-MY-SWEET-HOME')
```

```python
telefone('MY-MISERABLE-JOB')
```

```python
telefone('1J-WY1DV-LM01')
```

4- Dada um texto em formato string, encontre todas as palavras pertencentes a esta string que contenham uma determinada letra, passada como parâmetro. 

```python
import string
```

```python
def letraepalavra(letra, palavra):
    letra=letra.lower()
    palavra1=palavra.split()
    for word in palavra1:
        if letra in word.lower():
            print(word.strip(string.punctuation))
```
