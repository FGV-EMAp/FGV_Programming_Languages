```python
with open('../Dados/numero_mil_digitos.txt', mode='r') as f:
    numero = f.read()
```

```python
numero = numero.replace('\n','')
```

```python
print(numero)
```

```python
lista = [int(char) for char in list(numero)]
```

```python
%%time

i = 0
maior = 0
while i+13 <= 1000:
    temp = lista[i:i+13]
    mult = 1
    for elemento in temp:
        mult *= elemento
    if mult > maior:
        maior = mult
    i +=1
print(maior)
```
